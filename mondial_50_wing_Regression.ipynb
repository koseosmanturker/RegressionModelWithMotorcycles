```python
import csv
import re

with open('mondial_data.TXT') as f:
    data = f.read()

pattern = r'^(.*?)\t(\d{4})\t(\d+)\t(.*?)\t\n([\d.,]+) TL\n'
matches = re.findall(pattern, data, re.MULTILINE)

rows = []

for match in matches:
    name = match[0]
    year = int(match[1])
    km = int(match[2])
    color = match[3]
    price = match[4]
    price = price.replace(".", "")
    price = int(price)

    rows.append([name, year, km, color, price])

with open('extracted_data_mondial_50_wing.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Name', 'Year', 'Kilometers', 'Color', 'Price'])
    writer.writerows(rows)


```

```python
import pandas as pd

dataset = pd.read_csv("extracted_data_mondial_50_wing.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dataset["Color"] = labelencoder.fit_transform(dataset["Color"].values)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(dataset, test_size=0.3)
```

```python
X_train = train.drop(['Name', 'Price','Color'], axis=1)
Y_train = train.loc[:,'Price']

X_test = test.drop(['Name', 'Price', 'Color'], axis=1)
Y_test = test.loc[:,'Price']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from numpy import sqrt
```

```python
rmse = sqrt(mean_squared_error(Y_test, predictions))
rmse
```

```python
comparison = pd.DataFrame({"Actual Values": Y_test,"Predictions": predictions})
comparison.tail(10)
```
